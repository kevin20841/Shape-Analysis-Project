```python
# Import statements
import sys
sys.path.insert(0, '../')

import shapedist
from testing import examples as ex
import numpy as np
import tqdm
import scipy.interpolate
import matplotlib.pyplot as plt
import matplotlib
import scipy
matplotlib.rcParams['figure.figsize'] = [10, 8]
```

```python
# One example shapdist computation for bumpy gamma, uniform domain
n = 1024  # number of points in domain

t = np.linspace(0., 1., n)
q = ex.curve_example('limacon', t)[0]
p = np.zeros(q.shape)
x_function = scipy.interpolate.CubicSpline(t, q[0])
y_function = scipy.interpolate.CubicSpline(t, q[1])


test = ex.gamma_example("bumpy")[0]

p[0] = x_function(test(t))
p[1] = y_function(test(t))

p = p.T
q = q.T

energy, p_new, q_new, tg, gammay = shapedist.find_shapedist(p, q, 'ud', t1 = t, t2=t,shape_rep=shapedist.coords)
plt.plot(tg, gammay, ".b")
plt.plot(tg, test(tg), "-r")
plt.figure()
plt.ylim(-0.2, 0.2)
plt.xlim(-0.2, 0.2)
plt.plot(p_new[:, 0], p_new[:, 1], ".-b")
plt.figure()
plt.ylim(-0.2, 0.2)
plt.xlim(-0.2, 0.2)
plt.plot(q_new[:, 0], q_new[:, 1], ".-r")
plt.show()
```

```python
# One example shapdist compuation between two different shapes
n = 600 # number of points in domain 

t = np.linspace(0., 1., n)

p = ex.curve_example("flower", t)[0]
q = ex.curve_example("limacon", t)[0]


dist, p, q, tg, gammay= shapedist.find_shapedist(p.T, q.T, 'ud', t1 =t, t2=t)
inter_func1 = scipy.interpolate.CubicSpline(tg, q[:, 0])
inter_func2 = scipy.interpolate.CubicSpline(tg, q[:, 1])
q_reparam = np.array([inter_func1(gammay), inter_func2(gammay)])
q_reparam = q_reparam.T
plt.figure()
plt.plot(tg, gammay, ".-r")
plt.figure()
plt.plot(q_reparam[:, 0], q_reparam[:, 1], ".")
plt.plot(p[:, 0], p[:, 1], ".")
plt.plot([p[:, 0], q_reparam[:, 0]], [p[:, 1], q_reparam[:, 1]], "-y")
plt.show()
```
